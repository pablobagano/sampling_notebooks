This chapter focuses on Simple Random Sampling (SRS), a fundamental method in sampling theory. It explains when SRS is appropriate, how to calculate parameters, how to work with sampling with and without replacement, and how to estimate the population mean, total, and proportion. The chapter also covers variance estimation, standard errors, confidence intervals, and the comparison between SRS with replacement and SRS without replacement. ## <span style="color:red">3.1 SRS With and Without Replacement</span>

Key parameters:
- $N$: Population size
- $E_0$: Tolerable sampling error
- $n_0$: Approximate initial sample size
- $n$: Final sample size

Formula for initial sample size approximation:
$$
n_0 = \frac{1}{E_0^2}
$$

Then:
$$
n = \frac{N \cdot n_0}{N + n_0}
$$

**SRS With Replacement (AASc)**:  
Each selected unit is returned to the population before the next draw. This is rarely used in large-scale surveys but useful in theoretical contexts (e.g., multiple independent draws from the same distribution).

**SRS Without Replacement (AASs)**:  
Once a unit is selected, it is removed from the population. It cannot be selected again. This is commonly used in practice as it ensures unique elements in the sample.

**Example (With Replacement)**:  
Imagine repeatedly rolling a fair die. Each roll is an independent draw from the same distribution.

**Example (Without Replacement)**:  
If you have a population of 1,000 individuals, each uniquely numbered, and you select 220 for your sample, once an individual is chosen, they are not replaced, and the next draw comes from 999 remaining units.



## <span style="color:red">3.2 Description of the Sampling Plan</span>

A simple random sampling plan has four main steps. The difference between SRS with replacement (AASc) and without replacement (AASs) is primarily in the step concerning whether the chosen element is returned or not.

### With Replacement (AASc)
1. Enumerate all $N$ units: $\Omega = \{1, 2, \dots, N\}$.
2. Randomly select one unit, each with equal probability.
3. Return the chosen unit to $\Omega$.
4. Repeat the selection until $n$ units are chosen. Units can be repeated.

### Without Replacement (AASs)
1. Enumerate all $N$ units: $\Omega = \{1, 2, \dots, N\}$.
2. Randomly select one unit, each with equal probability.
3. Do not return the chosen unit. Now $\Omega$ has $N-1$ units.
4. Repeat until $ n $ unique units are chosen, without repetition.

**Example (AASc)**:  
$\Omega = \{1,2,3,4,5,6\}$. Drawing a number, replacing it, and continuing until you have the required sample size.

**Example (AASs)**:  
$\Omega = \{1,2,\dots,1000\}$. After choosing one number, say 462, it is not returned. For the next draw, the population effectively has 999 units.


## <span style="color:red">3.3 Estimators of the Mean, Total, and Population Proportion</span>

### With Replacement (AASc)

**Theorem 3.1 (Mean Estimator)**:
For a random variable $Y$ with $E(Y)=\mu$ and $V(Y)=\sigma^2$:
$$
\bar{y} = \frac{1}{n}\sum_{i=1}^{n} y_i
$$
is an unbiased and consistent estimator of \(\mu\), with:
$$
V(\bar{y}) = \frac{\sigma^2}{n}.
$$

**Total Estimator**:
If $\tau = \sum_{i=1}^{N}y_i$, then:
$$
\hat{\tau} = N \cdot \bar{y}, \quad V(\hat{\tau}) = N^2\frac{\sigma^2}{n}.
$$

**Proportion Estimator**:
For a binary variable $Y_i \in \{0,1\}$:
$$
P = \frac{1}{N}\sum_{i=1}^{N}Y_i, \quad \sigma^2 = P(1-P).
$$
The sample proportion:
$$
\hat{P} = \frac{1}{n}\sum_{i=1}^{n}Y_i
$$
is unbiased for $P$, with:
$$
V(\hat{P}) = \frac{P(1-P)}{n}.
$$

### Without Replacement (AASs)

**Theorem 3.3 (Mean and Total with AASs)**:
Mean is still unbiased:
$$
E(\bar{y}) = \mu, \quad V(\bar{y}) = \left(1-\frac{n}{N}\right)\frac{\sigma^2}{n}.
$$

For the total:
$$
\hat{\tau} = N\bar{y}, \quad E(\hat{\tau})=N\mu, \quad V(\hat{\tau}) = n\left(1-\frac{n}{N}\right)\sigma^2.
$$

For the proportion:
$$
E(\hat{P})=P, \quad V(\hat{P}) = \left(1-\frac{n}{N}\right)\frac{P(1-P)}{n}.
$$

The finite population correction $(1-n/N)$ reduces variance compared to the with-replacement scenario.


## <span style="color:red">3.4 Variance Estimator, Standard Error, and Confidence Interval (CI)</span>

### With Replacement (AASc)

Sample variance estimator:
$$
s^2 = \frac{1}{n-1}\sum_{i=1}^{n}(y_i - \bar{y})^2
$$
is unbiased for $\sigma^2$.

Standard error of the mean:
$$
EP(\bar{y}) = \frac{\sigma}{\sqrt{n}}.
$$

**Confidence Interval (for mean, assuming normality or large n)**:
$$
IC = \left(\bar{y} - Z_{0.975}\frac{\sigma}{\sqrt{n}}, \; \bar{y} + Z_{0.975}\frac{\sigma}{\sqrt{n}}\right)
$$

For a proportion:
$$
IC(P) = \left(\hat{P} - Z_{0.975}\frac{\sqrt{P(1-P)}}{\sqrt{n}}, \; \hat{P} + Z_{0.975}\frac{\sqrt{P(1-P)}}{\sqrt{n}}\right).
$$

### Without Replacement (AASs)

Include the finite population correction:
$$
EP(\bar{y}) = \sqrt{\left(1-\frac{n}{N}\right)\frac{\sigma^2}{n}}.
$$

Confidence Interval with correction:
$$
IC(\mu) = \left(\bar{y} - Z_{0.975}\sqrt{\left(1-\frac{n}{N}\right)\frac{\sigma^2}{n}}, \; \bar{y} + Z_{0.975}\sqrt{\left(1-\frac{n}{N}\right)\frac{\sigma^2}{n}}\right).
$$


## <span style="color:red">3.5 Comparing AASc with AASs by Estimator Variance</span>

Define the planning effect (EPA):
$$
EPA = \frac{\left(1-\frac{n}{N}\right)\frac{s^2}{n}}{\frac{\sigma^2}{n}} = \frac{N-n}{N-1}.
$$

For $n>1$, $\frac{N-n}{N-1}<1$, so $EPA<1$. This means SRS without replacement _(AASs)_ is more efficient than SRS with replacement _(AASc)_  because it reduces variance due to the finite population correction.

**Exercise 1: Estimating Mean Daily Rates of Hotel Rooms**

A European city has a population of 30 two-star hotels. Each hotel has a daily rate, and the goal is to estimate the average daily rate for a double room on vacation. Instead of checking all 30 hotels, we will draw a Simple Random Sample (SRS) of size 10.

- Population: 30 hotels with known daily rates (for example: 25, 20, 35, 21, 22, ...).
- Method: Write each hotel's rate on a slip of paper, mix them thoroughly, and randomly select 10 slips.
- Result (example of a random outcome): Suppose the sampled rates are {20, 24, 22, 28, 23, 24, 21, 20, 25, 27}.

From this sample, the sample mean can be computed as:
$$
\bar{y} = \frac{20 + 24 + 22 + 28 + 23 + 24 + 21 + 20 + 25 + 27}{10}
$$
This sample mean serves as an unbiased estimator of the population mean daily rate.

**Exercise 2: Determining Sample Size for a Desired Margin of Error**

Consider a population of size $N = 1000$ employees. We want to estimate the proportion who favor a certain training program. We desire a margin of error not exceeding 5%.

- Desired error: $E_0 = 0.05$
  
First, approximate the initial sample size:
$$
n_0 = \frac{1}{E_0^2} = \frac{1}{0.05^2} = 400
$$

Then adjust for the finite population:
$$
n = \frac{N \cdot n_0}{N + n_0} = \frac{1000 \times 400}{1000 + 400} = \frac{400000}{1400} \approx 285.7 \approx 285
$$

This calculation suggests that a sample of about 285 employees should achieve the desired margin of error, assuming a Simple Random Sample and certain conditions.

**Exercise 3: Estimating a Proportion of Elderly Residents in Households**

In a city with $N = 621$ households, we randomly select $n = 60$ households via SRS. Among these sampled households, $11$ have at least one resident aged 65 or older.

- Sample proportion:
$$
\hat{P} = \frac{11}{60} \approx 0.1833
$$

This sample proportion is an unbiased estimator of the true population proportion of such households. One might then consider the standard error and construct a confidence interval. To reduce a potential margin of error (for example, ensuring it is below 8 percentage points), one could recalculate the sample size following the same procedure as in Exercise 2.

**Exercise 4: Confidence Interval for Mean Number of Errors**

A shoe industry has 5 production units (referred to as "islands"): João (7 errors), Carlos (6), Tatiana (8), Pedro (6), and Roberta (5). We want a 95% confidence interval for the mean number of errors.

- Sample size: $n = 5$
- Sample mean:
$$
\bar{y} = \frac{7 + 6 + 8 + 6 + 5}{5} = 6.4
$$
- Sample standard deviation (approx.): $\sigma \approx 1.02$

For a 95% confidence interval:
$$
IC = \bar{y} \pm Z_{0.975}\frac{\sigma}{\sqrt{n}}
$$
Using $Z_{0.975} = 1.96$:
$$
IC \approx (6.4 - 1.96 \times \frac{1.02}{\sqrt{5}}, \; 6.4 + 1.96 \times \frac{1.02}{\sqrt{5}})
$$
Numerically, this interval might be approximately $(5.51; 7.29)$, indicating that with 95% confidence, the true mean number of errors per month lies within this range.

**Exercise 5: Confidence Interval for Mean Tree Diameter**

We measure the diameters (in inches) of 10 randomly selected chestnut trees of the same age in a certain forest:
19.4, 21.4, 22.3, 22.1, 20.1, 23.8, 24.6, 19.9, 21.5, 19.1.

- Sample mean:
$$
\bar{y} = \frac{19.4 + 21.4 + 22.3 + 22.1 + 20.1 + 23.8 + 24.6 + 19.9 + 21.5 + 19.1}{10} = 21.42
$$

- Sample standard deviation (approx.): $\sigma \approx 1.743$

For a 95% confidence interval for the mean:
$$
IC = \bar{y} \pm Z_{0.975}\frac{\sigma}{\sqrt{n}}
$$
$$
IC \approx (21.42 - 1.96 \times \frac{1.743}{\sqrt{10}}, \; 21.42 + 1.96 \times \frac{1.743}{\sqrt{10}})
$$
This yields approximately $(20.33; 22.5)$. We are 95% confident that the true mean diameter of such trees at 3 feet height lies between 20.33 and 22.5 inches.
